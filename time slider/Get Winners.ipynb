```python
import pandas as pd
df = pd.read_csv('countypres_2000-2016.csv')
```

```python
df.head()
```

```python
win = df.groupby(['FIPS', 'year'])['candidatevotes'].max()
dfw = win.to_frame()
dfw.head()
```

```python
df2 = dfw.merge(df, left_on=['FIPS', 'year'], right_on=['FIPS', 'year'])
df2.head()
```

```python
df3 = df2[df2['candidatevotes_x'] <= df2['candidatevotes_y']]
df3.head()
```

```python
df['FIPS'].nunique()
```

```python
len(df3)
```

```python
df3['FIPS'].nunique()
```

```python
winners = df3[['FIPS', 'year', 'candidate']]
winners.head()
```

```python
merged = df.merge(winners, left_on=['FIPS', 'year'], right_on=['FIPS', 'year'])
merged.head()
```

```python
df_rename = merged.rename(index=str, 
                   columns={
                       "candidate_x": "candidate",
                       "candidate_y": "winner"                   
                   })
```

```python
df_rename.to_csv("countypres_2000-2016_with_winner.csv")
```
